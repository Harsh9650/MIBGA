```python
import json
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def convergence(result_file, metric):
    results_per_algo = {}
    for i, j in result_file.items():
        per_test = {}
        for k, l in j.items():
            result_list = []
            for m in l.keys():
                if metric == "shortest": # taking shortest path at each generation.
                    result_list.append(min(l[m]))
                else:
                    result_list.append(np.mean(l[m])) # taking the average of best unique solutions generated at each generation
            per_test[k] = result_list
            
        results_per_algo[i] =  per_test       
    return results_per_algo


def convergence_plot(results_per_generation, state_name, metric_label):

    results_per_algo = convergence(results_per_generation, metric_label)

    # Create a figure with subplots
    fig, axs = plt.subplots(10, 5, figsize=(60, 80))  # Create a 10x5 grid of plots
    axs = axs.flatten()  # Flatten the array of axes for easy iteration

    # Set y-axis label based on the metric label
    y_lab = 'Path Length' if metric_label == "shortest" else 'Average Path Length (Top 10 solutions)'

    # Set file_name based on the metric label
    file_name = state_name+'_spc.jpg' if metric_label == "shortest" else state_name+'_t10c.jpg'

    for i, j in results_per_algo.items():
        for k, l in j.items():
            ax = axs[int(k)-1]
            ax.plot(results_per_algo['AHN'][k], label='AHN', marker='o', color='red')
            ax.plot(results_per_algo['MIBGA'][k], label='MIBGA', marker='x', color='blue')
            ax.set_xlabel(f'Test Case: {int(k)}', fontsize=30)
            ax.tick_params(axis='both', which='major', labelsize=20)
            ax.tick_params(axis='both', which='minor', labelsize=16)

    # Add global x and y labels
    fig.text(0.5, 0.03, 'Generation Count', ha='center', fontsize=100)
    fig.text(0.03, 0.5, y_lab, va='center', rotation='vertical', fontsize=100)

    # Create a global legend
    handles, labels = [], []
    for ax in axs:
        h, l = ax.get_legend_handles_labels()
        handles.extend(h)
        labels.extend(l)

    unique_labels = dict(zip(labels, handles))
    fig.legend(unique_labels.values(), unique_labels.keys(), loc='upper center', ncol=3, fontsize=50)
    
    plt.tight_layout(rect=[0.06, 0.06, 1, 0.95])  # Adjust layout to make room for the global labels

    plt.savefig(file_name, format='jpg')
    
    plt.show()
```

```python
# File paths and their corresponding variables
file_paths = {
    'ibga/Arizona/arizona_per_generation.json': 'arizona_per_generation',
    'ibga/Washington/washington_per_generation.json': 'washington_per_generation',
    'ibga/Kansas/kansas_per_generation.json': 'kansas_per_generation',
    'ibga/Puerto_Rico/puerto_rico_per_generation.json': 'puerto_rico_per_generation',
    'ibga/California/california_per_generation.json': 'california_per_generation',
    'ibga/Florida/florida_per_generation.json': 'florida_per_generation'
}

# Dictionary to hold the deserialized data
data = {}

# Loop through each file path and load the data
for file, var_name in file_paths.items():
    with open(file, 'r') as f:
        data[var_name] = json.load(f)

# Access the loaded data
arizona_per_generation = data['arizona_per_generation']
washington_per_generation = data['washington_per_generation']
kansas_per_generation = data['kansas_per_generation']
puerto_rico_per_generation = data['puerto_rico_per_generation']
california_per_generation = data['california_per_generation']
florida_per_generation = data['florida_per_generation']
```

# Convergence on the Florida Network

```python
convergence_plot(florida_per_generation, "florida", "shortest")
```

```python
convergence_plot(florida_per_generation, "florida", "average")
```
