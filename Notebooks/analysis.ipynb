```python
import pandas as pd
import numpy as np
from scipy.stats import wilcoxon
```

```python
def performance_count(result_data):
    mibga_count = 0
    equal_count = 0
    ahn_count = 0
    for i in range(len(result_data)):
        if result_data['MIBGA'][i] < result_data['AHN'][i]:
            mibga_count+=1
        elif result_data['MIBGA'][i] == result_data['AHN'][i]:
            equal_count += 1
        elif result_data['MIBGA'][i] > result_data['AHN'][i]:
            ahn_count += 1
    return mibga_count, equal_count, ahn_count
```

# Shortest Path

```python
arizona_sp = pd.read_excel("Arizona/arizona_sp.xlsx")
washington_sp = pd.read_excel("Washington/washington_sp.xlsx")
kansas_sp = pd.read_excel("Kansas/kansas_sp.xlsx")
puerto_rico_sp = pd.read_excel("Puerto_Rico/puerto_rico_sp.xlsx")
california_sp = pd.read_excel("California/california_sp.xlsx")
florida_sp = pd.read_excel("Florida/florida_sp.xlsx")


arizona_mibga_count, arizona_equal_count, arizona_ahn_count  = performance_count(arizona_sp)
washington_mibga_count, washington_equal_count, washington_ahn_count  = performance_count(washington_sp)
kansas_mibga_count, kansas_equal_count, kansas_ahn_count  = performance_count(kansas_sp)
puerto_rico_mibga_count, puerto_rico_equal_count, puerto_rico_ahn_count  = performance_count(puerto_rico_sp)
california_mibga_count, california_equal_count, california_ahn_count  = performance_count(california_sp)
florida_mibga_count, florida_equal_count, florida_ahn_count  = performance_count(florida_sp)
```

# Statistical Significance

```python
data1 = {
    'MIBGA_arizona': arizona_sp['MIBGA'],
    'MIBGA_washington': washington_sp['MIBGA'],
    'MIBGA_kansas': kansas_sp['MIBGA'],
    'MIBGA_puerto_rico': puerto_rico_sp['MIBGA'], 
    'MIBGA_california': california_sp['MIBGA'],
    'MIBGA_florida': florida_sp['MIBGA'],

    'C.AHN_arizona': arizona_sp['AHN'],
    'C.AHN_washington': washington_sp['AHN'],
    'C.AHN_kansas': kansas_sp['AHN'],
    'C.AHN_puerto_rico': puerto_rico_sp['AHN'], 
    'C.AHN_california': california_sp['AHN'],
    'C.AHN_florida': florida_sp['AHN']
}

shortest_path = pd.DataFrame(data1)

# Extract paired samples for each test
paired_tests1 = [
    (shortest_path['MIBGA_arizona'], shortest_path['C.AHN_arizona']),
    (shortest_path['MIBGA_washington'], shortest_path['C.AHN_washington']),
    (shortest_path['MIBGA_kansas'], shortest_path['C.AHN_kansas']),
    (shortest_path['MIBGA_puerto_rico'], shortest_path['C.AHN_puerto_rico']),
    (shortest_path['MIBGA_california'], shortest_path['C.AHN_california']),
    (shortest_path['MIBGA_florida'], shortest_path['C.AHN_florida'])]



# Perform the Wilcoxon signed-rank tests for the shortest path metric data.
p_values = []
for (sample1, sample2) in paired_tests1:
    stat, p_value = wilcoxon(sample1, sample2)
    p_values.append(p_value)

# Convert list to numpy array for easy manipulation
p_values = np.array(p_values)

# Number of tests
m = len(p_values)

# Original significance level
alpha = 0.05

# Bonferroni-corrected significance level
alpha_corrected = alpha / m

# Determine which p-values are significant after correction
significant = p_values < alpha_corrected

print("Original p-values:", p_values)
print("Bonferroni-corrected significance level:", alpha_corrected)
print("Significant after correction:", significant)
```
